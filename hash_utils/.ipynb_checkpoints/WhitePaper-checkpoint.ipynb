## Introduction

**HashJing** is a philosophical and aesthetic experiment at the intersection of generative art, cryptography, and Eastern symbolism. The name combines “hash” (a unique digital fingerprint) and the character **經** *jīng* — “canon” or “classical text.” In other words, it is a “Canon of the Hash”: a deterministic form where entropy becomes observable order.

Each 256-bit hash is transformed into a **mandala of 64 radial sectors and 4 concentric layers**: one hex symbol → one sector, one bit → one layer. This transformation of number into image raises questions: can randomness become meaningful? What if we view digital structures as modern ideograms? Where is the boundary between entropy and canon? References to the 64 hexagrams of the *I Ching* and the Daoist diagram of Wuji → Taiji connect this inquiry to an ancient tradition of contemplating change.

## 1. Хэш и случайность

В генеративной системе **HashJing** всё начинается с **256-битного хэша** — строки из 64 шестнадцатеричных символов с префиксом `0x`. Каждый такой хэш:

* **детерминирован** — один и тот же вход всегда даёт один и тот же выход;
* **выглядит случайным** — без знания входных данных его предсказать практически невозможно.

Эта двойственность делает хэш идеальным «семенем» визуальной формы: он надёжно фиксирует связь между числом и изображением, сохраняя при этом ореол загадки.

### Что такое хэш?
**Хэш-функция** — односторонняя математическая операция, переводящая сообщение произвольной длины в строку фиксированного размера. Например, так можно получить SHA-256 в Python:

```python
import hashlib
def hash_to_hex256(word: str) -> str:
    hash_bytes = hashlib.sha256(word.encode()).digest()
    return '0x' + hash_bytes.hex()
    
hash_to_hex256("HashJing")
# ➜ '0x2b054d225d025fc24c58030bda276b16089ae21cc8aff93d2751472a98eab06c'
```
#### Демонстрация функции генерации хэша из произвольного текста

````python
base_hash.hash_to_hex256("HashJing")
````

Полученные 64 hex-символа легко интерпретировать как 256 бит → **64 радиальных секторов × 4 слоя** в визуальном алгоритме HashJing (1 hex-символ → 1 сектор, 1 бит → 1 слой). Подробно о принципе построения бинарной хэш-мандалы будет рассказано в следующих разделах.

### Где брать хэши в блокчейне
| Источник                              | Применение в NFT-минте                                                   | Размер    |
| ------------------------------------- | ------------------------------------------------------------------------ | --------- |
| `tokenData.hash` (Art Blocks)         | Автоматически присваивается токену при минте                             | 256 бит (32 байта)  |
| `blockhash(n)`                        | Случайность на уровне блока, часто используют как `seed`                 | 256 бит (32 байта)  |
| `keccak256(tokenId ∥ secret)`         | Пользовательский способ получить непредсказуемый, но воспроизводимый хэш | 256 бит (32 байта)  |
| `sha256`, `blake2b`, Poseidon и т. д. | Вне-чейн генерация, если нужно доп. криптографическое качество           | переменно |

### Псевдослучайный хэш для тестов
Для набросков и отладки удобна простая функция:

 ```python
import random, string

def generate_random_hash() -> str:
    alphabet = string.hexdigits.lower()[:16]  # '0123456789abcdef'
    return '0x' + ''.join(random.choices(alphabet, k=64))
    
generate_random_hash()
# ➜ 0x7c3a...e1b4
```
⚠️ Это **не криптографическая** случайность (результат воспроизводим при фиксированном `random.seed()`), но для художественного прототипирования вполне достаточно.

#### Демонстрация функции генерации псевдослучайного хэша

### Детерминированная случайность как зеркало формы
Хэш служит **однозначным адресом** изображения и одновременно **загадкой**: обычнному человеку практически невозможно представить, какой именно узор породит строка, пока он не запустит генератор. Визуализация 256 бит делает видимыми симметрию, ритм и повторы, скрытые в чистом тексте — то самое «инь-ян цифровой энтропии», о котором говорит HashJing.

## 2. Бинарное представление числа и схемы «Книги Перемен»
«Книга Перемен» (*Ицзин*, 易 經) — пожалуй, первый известный пример **бинарной нотации** в истории.
Её «алфавит» — **гексаграмма**: вертикальная колонка из 6 линий, где

| Линия       | Символ | Бит | Семантика                   |
| ----------- | ------ | --- | --------------------------- |
| прерывистая | - -      | 0   | 陰 *инь* — тьма, пассивность |
| сплошная    | ---      | 1   | 陽 *ян* — свет, активность   |


Так каждая гексаграмма кодируется как число **от 0 до 63** (2⁶ – 1). Линии читаются **сверху вниз**; **младший** бит — верхняя черта, **старший** — нижняя.

| Уровень гексаграммы | 0 (000000) | 1 (000001) | 2 (000010) | … | 62 (111110) | 63 (111111) |
|:-:|:-:|:-:|:-:|:-:|:--:|:--:|
|⑥ | - - | --- | - - |   | - - | --- |
|⑤ | - - | - - | --- |   | --- | --- |
|④ | - - | - - | - - |   | --- | --- |
|③ | - - | - - | - - |   | --- | --- |
|② | - - | - - | - - |   | --- | --- |
|① | - - | - - | - - |   | --- | --- |

**Таблица вертикальной раскладки линий для гексаграмм в бинарном порядке.**
Строки ⑥ → ① — линии сверху (младший бит) к низу (старший бит); колонки приводят первые три и последние две гексаграммы полного ряда 0–63. `---` обозначает ян-черту (1), `- -` — инь-черту (0); пропуск «…» скрывает промежуточные столбцы.

<figure markdown>

</figure>

**Круг 64 гексаграмм в арифметическом порядке (неатрибутированная ксилография, PD).**
Стартовая точка — Кунь `000000` в верхней позиции. Двигаясь по часовой стрелке, мы каждый раз прибавляем `+1` к шестибитному числу: `000001`, `000010`, …, пока круг не замкнётся на Цянь `111111`. Чёрные ячейки = инь `0`, белые = ян `1`. Такая «нарастающая» раскладка совпадает с тем, как HashJing заполняет 64 сектора мандалы.

### От гексаграмм к HashJing

В *HashJing* мы берём ту же идею бинарной мономы, но применяем её к современному **256-битному хэшу** (SHA-256 / Keccak-256):

```
0x87a797a48cba94ee585ee2c0d7d6f4cce4dd12f77192a4d0bc562938d6fb62b1
```

* 64 hex-символа → **64 «сектора»** мандалы.
* Каждый символ = 4 бита → **4 концентрических слоя** (от центра к периферии).

Так возникает **естественный мост** между криптографической случайностью и древней бинарной космологией Ицзина: вместо шести линий мы читаем четыре, но логика «низший бит → внутренний слой» полностью наследует принципу «снизу вверх».

## 3. Принцип построения диаграмм

HashJing читает 256-битный хэш как **64 шестнадцатеричных символа**.
Каждый hex-символ → **4 бита** → 1 радиальный сектор с 4 концентрическими ячейками.

### ▸ Прямоугольная развёртка

Ниже приведена функция `draw_binary_grid_from_hex(my_hash)`, которая генерирует представление этой строки хэша в в виде ленты. Читается слева направо. Исходный код этой и других функций проекта можно посмотреть здесь.

### ▸ Круговая мандала

Ниже приведена функция `base_hash.draw_mandala(my_hash)`, которая генерирует круговую мандалу по по принципам, которые были описаны выше. Линейная лента показывает хэш «как есть», а мандала — тот же набор битов, свернутый в круг из 64 секторов и 4 слоёв. Такая трансформация делает видимой симметрию и «ритм» 256-битного числа, сохраняя при этом строгую детерминированность : один хэш → одна форма. 

* 64 сектора идут **по часовой стрелке**, старт — на «12 часов». Для наглядности демонстрации в начало отсчета добавлена красная радиальная линия.
* В каждом секторе ячейки от центра к периферии отображают биты `b0 → b1 → b2 → b3`.
* Белое = 1 (ян), чёрное = 0 (инь).
* В центр выводится сам хэш, разбитый на 4 строки × 16 символов — удобно сверять текст с изображением.

````python
my_hash = "0x87a797a48cba94ee585ee2c0d7d6f4cce4dd12f77192a4d0bc562938d6fb62b1"
base_hash.draw_mandala(my_hash, show_radial_line=True)
````

### ▸ Адреса как хэши: визуализация 160-битных форм

Помимо 256-битных хэшей, **HashJing** умеет превращать в мандалу и **Ethereum-адреса** — номера кошельков или смарт-контрактов. Внешне это всего лишь 40 hex-символов, но по сути — укороченный фрагмент Keccak-256-хэша длиной **160 бит**.

Поскольку `base_hash.draw_mandala()` принимает любую кратную-четырём длину, для адреса мы просто указываем число секторов = 40:

````python
base_hash.draw_mandala(
    "0xb47e3cd837ddf8e4c57f05d70ab865de6e193bbb",  # адрес CryptoPunks
    sectors=40, show_radial_line=True
)
````

Алгоритм остаётся тем же: каждый hex-символ даёт четыре концентрических ячейки, поэтому адрес проявляется как **40 радиальных секторов × 4 кольца**. Это позволяет «подписывать» мандалой любой объект сети — будь то кошелёк, контракт или NFT — сохраняя главный принцип HashJing: **один хэш → одна уникальная форма**.
